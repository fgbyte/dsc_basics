```python
# importamos el dataframe futbolistas para seguir trabajando con el
import pandas as pd
futbolistas = pd.read_excel(
    '../db/futbolistas.xlsx', index_col='Id', parse_dates=['Nacimiento'])
futbolistas.head()
```

```python
# hacemos las importaciones pertinentes
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

Crea un diagrama de cajas utilizando `boxplot` para mostrar la distribución de los precios en funcion de si los futbolistas asisten al gimnasio o no.

```python
sns.boxplot(data=futbolistas, x='Gimnasio', y='Precio')

# plt se integra al chart de seaborn porque corre por debajo
plt.title('Precio de los futbolistas por gimnasio')

plt.xlabel('Futbolista usa Gimnasio')
plt.ylabel('Precio')
plt.show()
```

Crea una grafica de puntos con jitter para mostrar la distribucion de la altura por liga utilizando `stripplot`.

Las graficas de jitter son para ver scatters(dispersiones) chicos agrupados por diferentes categorias.

```python
sns.stripplot(data=futbolistas, x='Liga', y='Altura', jitter=True)
plt.title('Distribución de alturas de futbolistas por liga')
plt.show()
```

Calcula la matriz de correlacion de las columnas numericas del DataFrame utilizando `corr`

```python
corr = futbolistas.corr(numeric_only=True)
corr
```

Crea un heatmap de la matriz corr con anotaciones, utilizando una paleta de colores 'coolwarm' centrada en 0 y con lineas de separaciones de 0.5.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, linewidths=.5)
plt.title('Heatmap de matriz de correlacion')
plt.show()
```
