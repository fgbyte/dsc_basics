Usaremos el mismo diccionario del ejemplo pasado para guardarlo y leerlo en los diferentes formatos compatibles

````python
import numpy as np
import pandas as pd
pd.__version__
````

````python


dict_data = {
    'edad': [10, 9, 13, 14, 12, 11, 12],
    'cm': [115, 110, 130, 155, 125, 120, 125],
    'pais': ['co', 'mx', 'co', 'mx', 'mx', 'ch', 'ch'],
    'genero': ['M', 'F', 'F', 'M', 'M', 'M', 'F'],
    'Q1': [5, 10, 8, np.nan, 2, 4, 3],
    'Q2': [7, 9, 9, 8, 8, 8, 9]
}

df = pd.DataFrame(dict_data)
df = pd.DataFrame(dict_data, index=[
                  'ana', 'benito', 'camilo', 'daniel', 'erika', 'fabian', 'gabriela'])
df
````

### Guardar a CSV

````python
# Guardar df a csv

path = '/home/fgbyte/study/dsc/basics/db/'

df.to_csv(path + 'data.csv')
# para no agregar index usamos el param index=False
````

### Leer de CSV

````python
df_read = pd.read_csv(path + 'data.csv', index_col=0)

# index_col=0 => el index va a ser la 1ra columna

df_read
````

### Guardar a Excel

````python
df.to_excel(path + 'data.xlsx', sheet_name='Hoja1')

# sheet_name Hoja1 => especifica el nombre de la hoja
````

### Leer de Excel âœ¨

````python
df_read = pd.read_excel(path + 'data.xlsx', sheet_name='Hoja1', index_col=0)

# sheet_name => especifica la hoja que queremos importar

df_read
````

### Guardar a JSON ðŸ’›

````python
df.to_json(path + 'data.json')
````

### Leer de JSON

````python
df_read = pd.read_json(path + 'data.json')
df_read

# el formato json es mas fiable a la hora de mantener los index
# es tambiÃ©n el default para trabajar con APIs
````

### Guardar a formato Pickle

Pickle is a Python-specific format for serializing (saving) and deserializing (loading) Python objects, including Pandas DataFrames. It's useful for saving complex data structures because it preserves the object's type and structure.

````python
df.to_pickle(path + 'data.pkl')
````

### Leer de formato Pickle

````python
df_read = pd.read_pickle(path + 'data.pkl')
df_read
````

### Guardar en formato Parquet
